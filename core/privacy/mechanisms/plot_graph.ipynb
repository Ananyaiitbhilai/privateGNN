```python
import matplotlib.pyplot as plt

# Define the data
data = [
    [0.1, 1, 1, 76.95],
     [0.2, 1, 1, 74.92],
     [0.5, 1, 1, 69.77],
    [0.9695, 1, 1, 64.28],
    [1, 1, 1, 64.11],
    [2, 1, 1, 54.30],
    [3, 1, 1, 49.32],
    [4, 1, 1, 48.46]
]

# Extract the noise scales and accuracy into separate lists
em_noise_scales = [x[0] for x in data]
am_noise_scales = [x[1] for x in data]
cm_noise_scales = [x[2] for x in data]
accuracy = [x[3] for x in data]

# Create a scatter plot with different symbols for different values of em_noise_scales
plt.scatter(am_noise_scales, accuracy, c=em_noise_scales, cmap='viridis', marker='o')

# Add axis labels and a color bar legend
plt.xlabel('am noise scale')
plt.ylabel('accuracy')
plt.colorbar(label='em noise scale')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Define the data
data = [
    [1, 1, 1, 64.11],
    [2, 1, 1, 54.30],
    [3, 1, 1, 49.32],
    [0.5, 1, 1, 69.77],
    [0.2, 1, 1, 74.92],
    [0.1, 1, 1, 76.95],
    [0.9695, 1, 1, 64.28],
    
    [4, 1, 1, 48.46]
]

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['em_noise', 'am_noise', 'cm_noise', 'accuracy'])

# Pivot the DataFrame to create a matrix of accuracy values
heatmap_data = df.pivot('am_noise', 'em_noise', 'accuracy')

# Create the heatmap using Seaborn
sns.heatmap(heatmap_data, cmap='coolwarm')

# Add axis labels
plt.xlabel('em noise scale')
plt.ylabel('am noise scale')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Define the data
data = [
   [1, 1, 1, 64.11],
    [2, 1, 1, 54.30],
    [3, 1, 1, 49.32],
    [0.5, 1, 1, 69.77],
    [0.2, 1, 1, 74.92],
    [0.1, 1, 1, 76.95],
    [0.9695, 1, 1, 64.28],
    [3, 1, 1, 49.32],
    [4, 1, 1, 48.46] 
]

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['em_noise', 'am_noise', 'cm_noise', 'accuracy'])

# Pivot the DataFrame to create a matrix of accuracy values
heatmap_data = df.pivot_table(values='accuracy', index='am_noise', columns=['em_noise', 'cm_noise'])

# Create the heatmap using Seaborn
sns.heatmap(heatmap_data, cmap='coolwarm')

# Add axis labels
plt.xlabel('em and cm noise scale')
plt.ylabel('am noise scale')

# Show the plot
plt.show()
```
