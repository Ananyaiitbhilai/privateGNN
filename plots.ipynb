```python
import pandas as pd
import plotly.graph_objects as go

# Load the CSV file into a pandas dataframe
df = pd.read_csv('table_red.csv')

# Add new columns for noise_am, noise_em, and noise_cm
df['noise_am'] = 1.0 * df['noise_scale']
df['noise_em'] = 1.0 * df['noise_scale']
df['noise_cm'] = 2.0 * df['noise_scale']

# Create a figure for Test Accuracy vs Epochs
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=df['epoch_number'], y=df['test_acc'], mode='lines', name='Test Accuracy'))
fig1.update_layout(title="Test Accuracy vs Epochs", height=500, width=700, font=dict(family='Arial', size=18),
                  xaxis=dict(title='Epochs'), yaxis=dict(title='Accuracy (%)', range=[80, 95]))
#fig1.show()

# Create a figure for Train Accuracy vs Epochs
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=df['epoch_number'], y=df['train_acc'], mode='lines', name='Train Accuracy'))
fig2.update_layout(title="Train Accuracy vs Epochs", height=500, width=700, font=dict(family='Arial', size=18),
                  xaxis=dict(title='Epochs'), yaxis=dict(title='Accuracy (%)', range=[80, 95]))
#fig2.show()

# Create a figure for Val Accuracy vs Epochs
fig3 = go.Figure()
fig3.add_trace(go.Scatter(x=df['epoch_number'], y=df['val_acc'], mode='lines', name='Val Accuracy'))
fig3.update_layout(title="Val Accuracy vs Epochs", height=500, width=700, font=dict(family='Arial', size=18),
                  xaxis=dict(title='Epochs'), yaxis=dict(title='Accuracy (%)', range=[80, 95]))
#fig3.show()

# Create a figure for Noise Scale per Module vs Epochs
fig4 = go.Figure()
fig4.add_trace(go.Scatter(x=df['epoch_number'], y=df['noise_am'], mode='lines', name='Noise Scale (AM)'))
fig4.add_trace(go.Scatter(x=df['epoch_number'], y=df['noise_em'], mode='lines', name='Noise Scale (EM)'))
fig4.add_trace(go.Scatter(x=df['epoch_number'], y=df['noise_cm'], mode='lines', name='Noise Scale (CM)'))
fig4.update_layout(title="Noise Scale per Module vs Epochs", height=500, width=700, font=dict(family='Arial', size=18),
                  xaxis=dict(title='Epochs'), yaxis=dict(title='Noise Scale', range=[0.9, 1.0]))
#fig4.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv("table_red.csv")

# Add columns for noise_am, noise_em, and noise_cm
df["noise_am"] = df["noise_scale"] * 1
df["noise_em"] = df["noise_scale"] * 1
df["noise_cm"] = df["noise_scale"] * 2

# Plot test accuracy vs epochs
plt.plot(df["epoch_number"], df["test_acc"])
plt.xlabel("Epoch Number")
plt.ylabel("Test Accuracy")
plt.title("Test Accuracy vs Epochs")
plt.show()

# Plot train accuracy vs epochs
plt.plot(df["epoch_number"], df["train_acc"])
plt.xlabel("Epoch Number")
plt.ylabel("Train Accuracy")
plt.title("Train Accuracy vs Epochs")
plt.show()

# Plot validation accuracy vs epochs
plt.plot(df["epoch_number"], df["val_acc"])
plt.xlabel("Epoch Number")
plt.ylabel("Validation Accuracy")
plt.title("Validation Accuracy vs Epochs")
plt.show()

# Plot noise scale per module vs epochs
plt.plot(df["epoch_number"], df["noise_am"], label="Module A")
plt.plot(df["epoch_number"], df["noise_em"], label="Module E")
plt.plot(df["epoch_number"], df["noise_cm"], label="Module C")
plt.xlabel("Epoch Number")
plt.ylabel("Noise Scale")
plt.title("Noise Scale per Module vs Epochs")
plt.legend()
plt.show()
```
